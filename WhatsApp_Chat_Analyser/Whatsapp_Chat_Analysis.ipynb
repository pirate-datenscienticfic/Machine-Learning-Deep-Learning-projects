```python
import re
import pandas as pd
```

```python
f = open("G:\Projects\Machine-Learning-Deep-Learning-projects\Whatsup_Chat_Analysis\Data\WhatsApp Chat with Irshad Comp A.txt", "r", encoding="utf-8")
```

```python
data = f.read()
```

```python
print(type(data))
```

```python
pattern = "\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s\wm\s-\s"

pattern
```

```python
messages = re.split(pattern, data)[1:]
print("Total_message in chat: ", len(messages))
```

```python
messages
```

```python
dates = re.findall(pattern, data)

dates
```

```python
df = pd.DataFrame({'User_message': messages, 'Message_dates': dates})

df.head()
```

```python
## Convert message_date&Time

df["Message_dates"] = pd.to_datetime(df["Message_dates"], format='%d/%m/%y, %I:%M %p - ')

df.rename(columns={"Message_dates": "Date"}, inplace=True)

df.head()
```

```python
df.shape
```

```python
## Separate users and messages
users = []
messages = []

for msg in df["User_message"]:
    entry = re.split("([\w\W]+?):\s", msg)
    if entry[1:]: #User Name
        users.append(entry[1])
        messages.append(entry[2][:-1])
    else:
        users.append("group_notification")
        messages.append(entry[0])
        

df["user"] = users
df["messages"] = messages
df.drop(columns=["User_message"], inplace=True)

df.head()
```

```python
df["Year"] = df["Date"].dt.year

df.head()
```

```python
df["Months"] = df["Date"].dt.month_name()

df["Month"] = df["Date"].dt.month

df.head()
```

```python
df["Day"] = df["Date"].dt.day

df["Hour"] = df["Date"].dt.hour
df["Minute"] = df["Date"].dt.minute
```

```python
period = []

for hour in df[["day_name", "Hour"]]["Hour"]:
    if hour == 23:
        period.append(str(hour) + "-" + str("00"))
    elif hour == 0:
        period.append(str("00") + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df["period"] = period
```
